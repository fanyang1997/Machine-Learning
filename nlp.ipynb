```python
from collections import Counter
import jieba
jieba.load_userdict('userdict.txt')

# åˆ›å»ºåœç”¨è¯list
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r').readlines()]
    return stopwords

# å¯¹å¥å­è¿›è¡Œåˆ†è¯
def seg_sentence(sentence):
    sentence_seged = jieba.cut(sentence.strip())
    stopwords = stopwordslist('G:\\å“ˆå·¥å¤§åœç”¨è¯è¡¨.txt')  # è¿™é‡ŒåŠ è½½åœç”¨è¯çš„è·¯å¾„
    outstr = ''
    for word in sentence_seged:
        if word not in stopwords:
            if word != '\t':
                outstr += word
                outstr += " "
    return outstr


inputs = open('hebing_wenben\\wenben.txt', 'r') #åŠ è½½è¦å¤„ç†çš„æ–‡ä»¶çš„è·¯å¾„
outputs = open('output.txt', 'w') #åŠ è½½å¤„ç†åçš„æ–‡ä»¶è·¯å¾„
for line in inputs:
    line_seg = seg_sentence(line)  # è¿™é‡Œçš„è¿”å›å€¼æ˜¯å­—ç¬¦ä¸²
    outputs.write(line_seg)
outputs.close()
inputs.close()
# WordCount
with open('output.txt', 'r') as fr: #è¯»å…¥å·²ç»å»é™¤åœç”¨è¯çš„æ–‡ä»¶
    data = jieba.cut(fr.read())
data = dict(Counter(data))

with open('cipin.txt', 'w') as fw: #è¯»å…¥å­˜å‚¨wordcountçš„æ–‡ä»¶è·¯å¾„
    for k, v in data.items():
        fw.write('%s,%d\n' % (k, v))
```
